```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r'C:\Users\spars\OneDrive\Desktop\ADM final term/creditcard_2023.csv')

# Display the first few rows and summary statistics
print(data.head())
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Visualizing distributions - example: Histogram of 'Amount'
plt.figure(figsize=(8, 6))
sns.histplot(data['Amount'], kde=True)
plt.title('Histogram of Transaction Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()

# Visualizing relationships - example: Scatter plot between 'V1' and 'V2'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='V1', y='V2', data=data)
plt.title('Scatter plot between V1 and V2')
plt.show()

# Correlation heatmap for numerical columns
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
