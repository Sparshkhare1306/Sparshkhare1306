```python
import numpy as np

# Define a function to handle samples
def process_samples(samples):
    for sample in samples:
        if np.asanyarray(sample).size == 0:
            return BartlettResult(np.nan, np.nan)
        if np.asanyarray(sample).ndim > 1:
            raise ValueError('Samples must be one-dimensional.')
class BartlettResult:
    def __init__(self, value1, value2):
        self.value1 = value1
        self.value2 = value2
your_samples = [[1, 2, 3], [4, 5, 6], [], [7, 8, 9]]


result = process_samples(your_samples)
print(result)  
```

```python
def BoxM(*samples):
    import numpy as np
    
    g = len(samples) # the number of groups
    p = samples[0].shape[1] # the number of variables
    cov_matrix = [] # the list of covariance matrix
    freedom = [] # the list of degrees of freedom of each group 
    for i in range(g):
        cov_matrix.append(samples[i].cov()) # dataframe自带计算协方差矩阵的方法
        freedom.append(samples[i].shape[0]-1) # 样本数-1
    # 计算s_p以及后续需要用到的中间量
    s_p = 0
    total_freedom = 0
    temp = 0
    temp1 = 0
    temp2 = 0
    for i in range(g):
        s_p += freedom[i]*cov_matrix[i]
        total_freedom += freedom[i]
        temp += freedom[i]*np.log(np.linalg.det(cov_matrix[i]))
        temp1 += 1/freedom[i]
        temp2 += 1/(freedom[i]*freedom[i])
    s_p = s_p/total_freedom
    # print("s_p:",s_p)
    # 计算M
    M = total_freedom*np.log(np.linalg.det(s_p)) - temp
    # 本例中各组样本量>>20，p=29>5，g=2<5，适用于F近似
    A1 = (temp1-1/total_freedom)*(2*p*p+3*p-1)/(6*(p+1)*(g-1))
    A2 = (temp2-1/(total_freedom*total_freedom))*(p-1)*(p+2)/(6*(g-1))
    print("A1:",A1)
    print("A2:",A2)
    f1 = 1/2*(g-1)*p*(p+1)
    f2 = (f1+2)/(A2-A1*A1)
    b = f1/(1-A1-f1/f2)
    print("f1:",f1)
    print("f2:",f2)
    print("b:",b)
    from scipy import stats
    pvalue = stats.f.sf(M/b,f1,f2)
    return M/b, pvalue
print("Done")
Xtrain_0 = Xtrain[Xtrain["Class"]==0] #根据标签列获得两组的样本
Xtrain_1 = Xtrain[Xtrain["Class"]==1]
Xtrain_0 = Xtrain_0.drop('Class', axis=1) #删除标签列
Xtrain_1 = Xtrain_1.drop('Class', axis=1)
M_stastic, p_value = BoxM(Xtrain_0,Xtrain_1)
print(M_stastic)
print(p_value)
```

# Estimate the parameters of the conditional probability density functions and Compute discriminant functions

```python
class DiscriminationAnalysis():
    def __init__(self,prior):
        self.prior = prior
        self.s_0 = None
        self.mean_0 = None
        self.s_1 = None
        self.mean_1 = None
        self.classification_result = []
    def fit(self,X,Y):
        full = pd.concat([X,Y],axis=1,ignore_index=False)
        full_0 = full[full["Class"]==0] 
        full_1 = full[full["Class"]==1]
        X_0 = full_0.drop('Class', axis=1) 
        X_1 = full_1.drop('Class', axis=1)
        self.s_0 = X_0.cov()
        self.mean_0 = X_0.mean(axis=0)
        self.s_1 = X_1.cov()
        self.mean_1 = X_1.mean(axis=0)
    def classification(self,row):
        x = np.array(row)
        Q_score_0 = -1/2*np.log(np.linalg.det(self.s_0))-1/2*np.dot(np.dot((x-self.mean_0).T, np.linalg.inv(self.s_0)), (x-self.mean_0))+np.log(self.prior[0])
        Q_score_1 = -1/2*np.log(np.linalg.det(self.s_1))-1/2*np.dot(np.dot((x-self.mean_1).T, np.linalg.inv(self.s_1)), (x-self.mean_1))+np.log(self.prior[1])
        if Q_score_0 > Q_score_1:
            self.classification_result.append(0)
        else:
            self.classification_result.append(1)
    def score(self,X,Y):
        self.classification_result.clear()
        print(len(self.classification_result))
        X.apply(self.classification,axis=1)
        print(len(self.classification_result))
        right = 0
        for i in range(X.shape[0]):
            if self.classification_result[i] == Y.iloc[i]:
                right += 1
        return right/X.shape[0]
```

```python
DA = DiscriminationAnalysis(prior)
if "Class" in Xtrain.columns:
    Xtrain = Xtrain.drop("Class",axis=1)
full_train_valid = pd.concat([Xtrain,Ytrain],axis=1,ignore_index=False)
train_valid_0 = full_train_valid[full_train_valid["Class"]==0] 
train_valid_1 = full_train_valid[full_train_valid["Class"]==1]

accuracy_list =[]
num_0 = train_valid_0.shape[0]
num_1 = train_valid_1.shape[0]
for i in range(4):    
    sample_valid_0 = train_valid_0[int(num_0/4*i):int(num_0/4*(i+1))]
    sample1_0 = train_valid_0[:int(num_0/4*i)]
    sample2_0 = train_valid_0[int(num_0/4*(i+1)):]
    sample_train_0 = pd.concat([sample1_0,sample2_0],axis=0,ignore_index=False)
    sample_valid_1 = train_valid_1[int(num_1/4*i):int(num_1/4*(i+1))]
    sample1_1 = train_valid_1[:int(num_1/4*i)]
    sample2_1 = train_valid_1[int(num_1/4*(i+1)):]
    sample_train_1 = pd.concat([sample1_1,sample2_1],axis=0,ignore_index=False)
    
    sample_train_01 = pd.concat([sample_train_0,sample_train_1],axis=0,ignore_index=False) 
    sample_valid_01 = pd.concat([sample_valid_0,sample_valid_1],axis=0,ignore_index=False) 
    sample_train_y = sample_train_01["Class"]
    sample_train_x = sample_train_01.drop("Class",axis=1)
    sample_valid_y = sample_valid_01["Class"]
    sample_valid_x = sample_valid_01.drop("Class",axis=1)

    
    DA.fit(sample_train_x,sample_train_y)
    accuracy = DA.score(sample_valid_x,sample_valid_y)
    accuracy_list.append(accuracy)
print(accuracy_list)
```
